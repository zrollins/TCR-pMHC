```python
# Import hx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/100/hx.xlsx')
df2 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/95/hx.xlsx')
df3 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/90/hx.xlsx')

df4 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/100/hx.xlsx')
df5 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/95/hx.xlsx')
df6 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/90/hx.xlsx')

df7 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/100/hx.xlsx')
df8 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/95/hx.xlsx')
df9 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/90/hx.xlsx')

#Combine Data

df = df.append(df2, ignore_index=True)
df = df.append(df3, ignore_index=True)
print(df)

df4 = df4.append(df5, ignore_index=True)
df4 = df4.append(df6, ignore_index=True)

df7 = df7.append(df8, ignore_index=True)
df7 = df7.append(df9, ignore_index=True)

#Bin H-Bonds in Distance Bins
bins=np.linspace(5.884,14.612,200).tolist()

MHC = df.groupby(pd.cut(df['COM Distance (nm)'], bins=bins))['MHC_Hs'].agg(['mean','sem','size'])
pep = df.groupby(pd.cut(df['COM Distance (nm)'], bins=bins))['Peptide_Hs'].agg(['mean','sem','size'])

MHC4 = df4.groupby(pd.cut(df4['COM Distance (nm)'], bins=bins))['MHC_Hs'].agg(['mean','sem','size'])
pep4 = df4.groupby(pd.cut(df4['COM Distance (nm)'], bins=bins))['Peptide_Hs'].agg(['mean','sem','size'])

MHC7 = df7.groupby(pd.cut(df7['COM Distance (nm)'], bins=bins))['MHC_Hs'].agg(['mean','sem','size'])
pep7 = df7.groupby(pd.cut(df7['COM Distance (nm)'], bins=bins))['Peptide_Hs'].agg(['mean','sem','size'])

# Plot H-Bonds vs Distance

ax = plt.gca()
MHC.plot(kind='line', y='mean',yerr='sem', color='magenta', use_index=True, ax=ax)
MHC4.plot(kind='line',y='mean',yerr='sem', color='orange', use_index=True,ax=ax)
MHC7.plot(kind='line',y='mean',yerr='sem', color='purple', use_index=True,ax=ax)
ax.set_ylabel('Number of H-Bonds',fontname = 'Arial', fontsize=20)
ax.set_xlabel('Reaction Coordinate,  (nm)',fontname = 'Arial', fontsize=20)
xticks=bins[::20]
xticks=np.around(xticks,1)
plt.xticks(range(0,len(MHC),20), xticks)
ax.legend(['MART1(WT)','GVA','L1'])
ax.set_title('TCR-MHC H-Bond Distance',fontname = 'Arial Black', fontsize=20)
plt.xlim((0,75))
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.savefig('/Users/zrollins/Box/DMF5_MART1/hx_mhc.png', bbox_inches='tight',dpi=300)
plt.show()

ax = plt.gca()
pep.plot(kind='line', y='mean',yerr='sem', color='magenta', ax=ax)
pep4.plot(kind='line', y='mean',yerr='sem', color='orange', ax=ax)
pep7.plot(kind='line', y='mean',yerr='sem', color='purple', ax=ax)
ax.set_ylabel('Number of H-Bonds',fontname = 'Arial', fontsize=20)
ax.set_xlabel('Reaction Coordinate,  (nm)',fontname = 'Arial', fontsize=20)
plt.xticks(range(0,len(MHC),20), xticks)
ax.legend(['MART1(WT)','GVA','L1'])
ax.set_title('TCR-Peptide H-Bond Distance',fontname = 'Arial Black', fontsize=20)
plt.xlim((0,75))
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.savefig('/Users/zrollins/Box/DMF5_MART1/hx_pep.png', bbox_inches='tight',dpi=300)
plt.show()
```
