```python
# SASA Bins
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

t, t1, t2, t3, t4, t5, t6, t7, t8, pep, pep1, pep2, pep3, pep4, pep5, pep6, pep7, pep8 = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]
# 100 ns MART1
with open("/Users/zrollins/Box/DMF5_MART1/M1+/100/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t.append(float(cols[0]))
            pep.append(float(cols[1]))
# 95 ns MART1
with open("/Users/zrollins/Box/DMF5_MART1/M1+/95/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t1.append(float(cols[0]))
            pep1.append(float(cols[1]))
# 90 ns MART1 
with open("/Users/zrollins/Box/DMF5_MART1/M1+/90/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t2.append(float(cols[0]))
            pep2.append(float(cols[1]))
# 100 ns GVA
with open("/Users/zrollins/Box/DMF5_MART1/GVA+/100/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t3.append(float(cols[0]))
            pep3.append(float(cols[1]))
# 95 ns GVA
with open("/Users/zrollins/Box/DMF5_MART1/GVA+/95/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t4.append(float(cols[0]))
            pep4.append(float(cols[1]))
# 90 ns GVA
with open("/Users/zrollins/Box/DMF5_MART1/GVA+/90/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t5.append(float(cols[0]))
            pep5.append(float(cols[1]))

# 100 ns L1
with open("/Users/zrollins/Box/DMF5_MART1/L1+/100/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t6.append(float(cols[0]))
            pep6.append(float(cols[1]))

# 95 ns L1
with open("/Users/zrollins/Box/DMF5_MART1/L1+/95/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t7.append(float(cols[0]))
            pep7.append(float(cols[1]))
            
#90 ns L1 
with open("/Users/zrollins/Box/DMF5_MART1/L1+/90/SASA_pep.xvg") as f:
    for line in f:
        if line.startswith('#'):
            continue
        if line.startswith('@'):
            continue
        cols = line.split()

        if len(cols) == 2:
            t7.append(float(cols[0]))
            pep7.append(float(cols[1]))


print(np.amin(pep))
print(np.amax(pep))

#Bin SASA  
cut_bins = [12.0,12.1,12.2,12.3,12.4,12.5,12.6,12.7,12.8,12.9,13.0,13.1,13.2,13.3,13.4,13.5,13.6,13.7,13.8,13.9,14.0,14.1,14.2]

sasa={'MART1 (100ns)': pep}
df = pd.DataFrame(sasa)
s=pd.cut(df['MART1 (100ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa1={'MART1 (95ns)': pep1}
df1 = pd.DataFrame(sasa1)
s1=pd.cut(df1['MART1 (95ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa2={'MART1 (90ns)': pep2}
df2 = pd.DataFrame(sasa2)
s2=pd.cut(df2['MART1 (90ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa3={'GVA (100ns)': pep3}
df3 = pd.DataFrame(sasa3)
s3=pd.cut(df3['GVA (100ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa4={'GVA (95ns)': pep4}
df4 = pd.DataFrame(sasa4)
s4=pd.cut(df4['GVA (95ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa5={'GVA (90ns)': pep5}
df5 = pd.DataFrame(sasa5)
s5=pd.cut(df5['GVA (90ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa6={'L1 (100ns)': pep6}
df6 = pd.DataFrame(sasa6)
s6=pd.cut(df6['L1 (100ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa7={'L1 (95ns)': pep7}
df7 = pd.DataFrame(sasa7)
s7=pd.cut(df7['L1 (95ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

sasa8={'L1 (90ns)': pep8}
df8 = pd.DataFrame(sasa8)
s8=pd.cut(df8['L1 (90ns)'], bins=cut_bins).value_counts(sort=False, normalize=True)

#Plot SASA Frequency

SASA = s.to_frame()
s1 = s1.to_frame()
SASA['MART1 (95ns)'] = s1['MART1 (95ns)']
s2 = s2.to_frame()
SASA['MART1 (90ns)'] = s2['MART1 (90ns)']

s3 = s3.to_frame()
SASA['GVA (100ns)'] = s3['GVA (100ns)']
s4 = s4.to_frame()
SASA['GVA (95ns)'] = s4['GVA (95ns)']
s5 = s5.to_frame()
SASA['GVA (90ns)'] = s5['GVA (90ns)']

s6 = s6.to_frame()
SASA['L1 (100ns)'] = s6['L1 (100ns)']
s7 = s7.to_frame()
SASA['L1 (95ns)'] = s7['L1 (95ns)']
s8 = s8.to_frame()
SASA['L1 (90ns)'] = s8['L1 (90ns)']

l = [12.0,12.1,12.2,12.3,12.4,12.5,12.6,12.7,12.8,12.9,13.0,13.1,13.2,13.3,13.4,13.5,13.6,13.7,13.8,13.9,14.0,14.1]
x = pd.Series(l)
SASA['SASA']= x.values

M1 = ['MART1 (100ns)','MART1 (95ns)','MART1 (90ns)']
SASA['MART1 (WT)'] = SASA[M1].astype(float).mean(axis=1)
SASA['M1 SEM'] = SASA[M1].astype(float).sem(axis=1)

GVA = ['GVA (100ns)','GVA (95ns)','GVA (90ns)']
SASA['GVA'] = SASA[GVA].astype(float).mean(axis=1)
SASA['GVA SEM'] = SASA[GVA].astype(float).sem(axis=1)

L1 = ['L1 (100ns)','L1 (95ns)','L1 (90ns)']
SASA['L1'] = SASA[L1].astype(float).mean(axis=1)
SASA['L1 SEM'] = SASA[L1].astype(float).sem(axis=1)

ax =plt.gca()
SASA.plot(kind='line',x='SASA', y='MART1 (WT)', yerr = 'M1 SEM', color='magenta', ax=ax)
SASA.plot(kind='line',x='SASA', y='GVA', yerr = 'GVA SEM', color='orange', ax=ax)
SASA.plot(kind='line',x='SASA', y='L1', yerr ='L1 SEM',color='purple', ax=ax)
ax.set_ylabel(r'Probability Density ($\frac{N*s^2}{nm}$)',fontname= 'Arial', fontsize=20)
#ax.set_ylabel(r'Probability Density ($\frac{N*s^2}{nm}$)',fontname= 'Arial', fontsize=20)
ax.set_xlabel( r'Hydrophobic Surace Area ($\frac{nm}{N*s^2}$)', fontname= 'Arial', fontsize=20)
#ax.set_title('Peptide Solvent Accessible Surface Area', fontname='Arial', fontsize='20')
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.savefig('/Users/zrollins/Box/DMF5_MART1/SASAtot.png',figsize=(6,4),bbox_inches='tight',dpi=300)
plt.show()
```
