```python
#H-Frequency
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/hfrequency.xlsx')
df2 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/100/hfrequency.xlsx')
df3 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/95/hfrequency.xlsx')
df4 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/M1+/90/hfrequency.xlsx')

df5 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/hfrequency.xlsx')
df6 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/100/hfrequency.xlsx')
df7 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/95/hfrequency.xlsx')
df8 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/GVA+/90/hfrequency.xlsx')

df9 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/hfrequency.xlsx')
df10 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/100/hfrequency.xlsx')
df11 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/95/hfrequency.xlsx')
df12 = pd.read_excel (r'/Users/zrollins/Box/DMF5_MART1/L1+/90/hfrequency.xlsx')


#MART1

df['100 ns H-Bond Frequency (pMHC)'] = pd.Series(df2['H-Bond Frequency (pMHC)'])
df['100 ns Total Time (ps)'] = pd.Series(df2['Total Time (ps)'])

df['95 ns H-Bond Frequency (pMHC)'] = pd.Series(df3['H-Bond Frequency (pMHC)'])
df['95 ns Total Time (ps)'] = pd.Series(df3['Total Time (ps)'])

df['90 ns H-Bond Frequency (pMHC)'] = pd.Series(df4['H-Bond Frequency (pMHC)'])
df['90 ns Total Time (ps)'] = pd.Series(df4['Total Time (ps)'])

#GVA

df5['100 ns H-Bond Frequency (pMHC)'] = pd.Series(df6['H-Bond Frequency (pMHC)'])
df5['100 ns Total Time (ps)'] = pd.Series(df6['Total Time (ps)'])

df5['95 ns H-Bond Frequency (pMHC)'] = pd.Series(df7['H-Bond Frequency (pMHC)'])
df5['95 ns Total Time (ps)'] = pd.Series(df7['Total Time (ps)'])

df5['90 ns H-Bond Frequency (pMHC)'] = pd.Series(df8['H-Bond Frequency (pMHC)'])
df5['90 ns Total Time (ps)'] = pd.Series(df8['Total Time (ps)'])

#L1

df9['100 ns H-Bond Frequency (pMHC)'] = pd.Series(df10['H-Bond Frequency (pMHC)'])
df9['100 ns Total Time (ps)'] = pd.Series(df10['Total Time (ps)'])

df9['95 ns H-Bond Frequency (pMHC)'] = pd.Series(df11['H-Bond Frequency (pMHC)'])
df9['95 ns Total Time (ps)'] = pd.Series(df11['Total Time (ps)'])

df9['90 ns H-Bond Frequency (pMHC)'] = pd.Series(df12['H-Bond Frequency (pMHC)'])
df9['90 ns Total Time (ps)'] = pd.Series(df12['Total Time (ps)'])

df.fillna(0, inplace=True)
df5.fillna(0, inplace=True)
df9.fillna(0, inplace=True)

cols = ['100 ns H-Bond Frequency (pMHC)','95 ns H-Bond Frequency (pMHC)','90 ns H-Bond Frequency (pMHC)']
df['500 pN Frequency'] = df[cols].astype(float).mean(axis=1)
df['H-Bond Frequency SEM'] = df[cols].astype(float).sem(axis=1)

df5['500 pN Frequency'] = df5[cols].astype(float).mean(axis=1)
df5['H-Bond Frequency SEM'] = df5[cols].astype(float).sem(axis=1)

df9['500 pN Frequency'] = df9[cols].astype(float).mean(axis=1)
df9['H-Bond Frequency SEM'] = df9[cols].astype(float).sem(axis=1)

cols2 = ['100 ns Total Time (ps)','95 ns Total Time (ps)','90 ns Total Time (ps)']
df['500 pN Total Time'] = df[cols2].astype(float).mean(axis=1)
df['Total Time SEM'] = df[cols2].astype(float).sem(axis=1)

df5['500 pN Total Time'] = df5[cols2].astype(float).mean(axis=1)
df5['Total Time SEM'] = df5[cols2].astype(float).sem(axis=1)

df9['500 pN Total Time'] = df9[cols2].astype(float).mean(axis=1)
df9['Total Time SEM'] = df9[cols2].astype(float).sem(axis=1)


#print(df)
#print(df5)
#print(df9)

ax = plt.gca()
#df.plot(kind='line',x='Total_Hs', y='H-Bond Frequency (pMHC)', color='black', ax=ax)
df.plot(kind='line',x='Total_Hs', y='500 pN Frequency', yerr = 'H-Bond Frequency SEM', color='magenta', ax=ax)
#df5.plot(kind='line',x='Total_Hs', y='H-Bond Frequency (pMHC)', color='black', ax=ax)
df5.plot(kind='line',x='Total_Hs', y='500 pN Frequency', yerr = 'H-Bond Frequency SEM', color='orange', ax=ax)
#df9.plot(kind='line',x='Total_Hs', y='H-Bond Frequency (pMHC)', color='black', ax=ax)
df9.plot(kind='line',x='Total_Hs', y='500 pN Frequency', yerr = 'H-Bond Frequency SEM', color='purple', ax=ax)
ax.set_ylabel(r'Probability Density ($\frac{1}{Number}$)', fontname='Arial', fontsize='20')
ax.set_xlabel('Number of Hydrogen Bonds', fontname='Arial', fontsize='20')
ax.legend(['MART1(WT)','GVA','L1'])
#ax.legend(['EQ','GVA'])
#ax.set_title('TCR-pMHC Hydrogen Bonding', fontname='Arial', fontsize='20')
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.savefig('/Users/zrollins/Box/DMF5_MART1/hfreq.png', figsize=(6,4),bbox_inches='tight',dpi=300)
plt.show()


#ax = plt.gca()
#df.plot(kind='line',x='Total_Hs', y='Total Time (ps)', color='black', ax=ax)
#df.plot(kind='line',x='Total_Hs', y='500 pN Total Time', yerr = 'Total Time SEM', color='magenta', ax=ax)
#df5.plot(kind='line',x='Total_Hs', y='Total Time (ps)', color='grey', ax=ax)
#df5.plot(kind='line',x='Total_Hs', y='500 pN Total Time', yerr = 'Total Time SEM', color='orange', ax=ax)
#df9.plot(kind='line',x='Total_Hs', y='500 pN Total Time', yerr = 'Total Time SEM', color='purple', ax=ax)
#ax.set_xlabel('Number of Hydrogen Bonds')
#ax.set_ylabel('H-Bond Relative Time (ps)')
#ax.legend(['MART1 (WT)','GVA','L1'])
#ax.set_title('DMF5-MART1 H-Bond Relative Time')
#plt.savefig('/Users/zrollins/Box/DMF5_MART1/hrtime.png')
#plt.show()
```
